**Cargar el Dataset de Iris de manera virtual**

```python
import pandas as pd

# Cargar el dataset de Iris desde una URL pública
iris_url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'
iris_columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species']
iris_data = pd.read_csv(iris_url, header=None, names=iris_columns)
iris_data.head()

```

**Crear gráficos de dispersión y de distribución usando Matplotlib**

```python
import matplotlib.pyplot as plt

# Gráfico de dispersión para longitud y ancho de sépalo y pétalo
plt.figure(figsize=(8, 6))
plt.scatter(iris_data['sepal_length'], iris_data['sepal_width'], c='blue', label='Sepal')
plt.scatter(iris_data['petal_length'], iris_data['petal_width'], c='red', label='Petal')
plt.xlabel('Length')
plt.ylabel('Width')
plt.title('Scatter Plot de Sepal vs Petal')
plt.legend()
plt.show()

# Histograma para la distribución de la longitud del sépalo
plt.figure(figsize=(8, 6))
plt.hist(iris_data['sepal_length'], bins=20, color='purple', alpha=0.7)
plt.xlabel('Sepal Length')
plt.ylabel('Frequency')
plt.title('Distribución de Longitud de Sépalo')
plt.show()
```

**Crear gráficos de pares y mapas de calor usando Seaborn**

```python
import seaborn as sns

# Gráfico de pares para observar relaciones entre todas las variables, sin incluir la columna de especies en la correlación
sns.pairplot(iris_data, hue='species', palette='husl')
plt.show()

# Mapa de calor de correlación entre las variables numéricas, excluyendo la columna 'species'
plt.figure(figsize=(10, 8))
sns.heatmap(iris_data.drop('species', axis=1).corr(), annot=True, cmap='coolwarm')
plt.title('Mapa de Calor de Correlación')
plt.show()

```

**Comparar la visualización de datos en Ciencia de Datos vs. Big Data usando el Dataset de Titanic**

```python
# Cargar el dataset de Titanic desde una URL pública
titanic_url = 'https://web.stanford.edu/class/archive/cs/cs109/cs109.1166/stuff/titanic.csv'
titanic_data = pd.read_csv(titanic_url)
titanic_data.head()
```

**Crear visualizaciones con diferentes aspectos del Dataset de Titanic (Pyplot y Seaborn)**

```python
# Supervivencia por clase
plt.figure(figsize=(8, 6))
sns.countplot(x='Pclass', hue='Survived', data=titanic_data, palette='pastel')
plt.title('Supervivencia por Clase')
plt.xlabel('Clase')
plt.ylabel('Conteo')
plt.legend(title='Sobrevivió', labels=['No', 'Sí'])
plt.show()

# Distribución de Edad por Supervivencia
plt.figure(figsize=(10, 6))
sns.histplot(titanic_data[titanic_data['Survived'] == 1]['Age'], bins=30, color='green', kde=True)
sns.histplot(titanic_data[titanic_data['Survived'] == 0]['Age'], bins=30, color='red', kde=True)
plt.title('Distribución de Edad por Supervivencia')
plt.xlabel('Edad')
plt.legend(['Sobrevivió', 'No sobrevivió'])
plt.show()

# Supervivencia por género
plt.figure(figsize=(8, 6))
sns.countplot(x='Sex', hue='Survived', data=titanic_data, palette='muted')
plt.title('Supervivencia por Género')
plt.xlabel('Género')
plt.ylabel('Conteo')
plt.legend(title='Sobrevivió', labels=['No', 'Sí'])
plt.show()
```
